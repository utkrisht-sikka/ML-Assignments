```python
idd = 0
meanrgb, stdrgb = [0.485, 0.456, 0.406], [0.229, 0.224, 0.225]


import torch
import torch.nn as nn

from torch.utils.data import  random_split as slices
# dataloader takes the train set and is a abstraction over a class that provides batches of the dataset when used in a for loop. It 
# breaks the data set into batches of mentioned batchsize.
# random_split is used to split data into training , validation,test set
import torchvision as biglight
#T will be used to apply transformations like resizing image, normalizing, greyscale conversion etc.
from torchvision.datasets import FashionMNIST
# we will  be downloading FashionMNIST data from amazon
from torch.nn import Sequential as bunchoflayers
from torchvision.models import vgg16 as inbuiltmodel
import pickle, gc
# nn stands for neural network. It provides activation layers, cnn layers, linear layers etc.
from torch.utils.data import DataLoader as abcd
import numpy as np
# with this we import numpy module
import pandas as pd
# with this we import pandas module
collectiveset = 100
random_seed = 0
torch.manual_seed(random_seed)
applychanges = biglight.transforms.Compose([biglight.transforms.Resize(size=224), biglight.transforms.Grayscale(num_output_channels=3), biglight.transforms.ToTensor(), biglight.transforms.Normalize(meanrgb, stdrgb)])
# biglight.transforms.Resize resizes the image to 224X224
# biglight.transforms.Grayscale converts the image to greyscale
# biglight.transforms.ToTensor converts the image/np array to tensor
# biglight.transforms.Normalize normalizes the images using mean and std. of imagenet dataset. 3 means because of 3 colors: R,G,B

directory = './data'
tplusv = FashionMNIST(root=directory, download=True, transform=applychanges)

```

```python
def convert_prob_to_exactvalues(generated, labels):
    useless2, yhat = torch.max(generated, dim=1)
    correct_tensor = yhat.eq(labels.data.view_as(yhat))
             
    return torch.mean(correct_tensor.type(torch.FloatTensor)).item()
```

```python
def fitthemodel(model, lossfunc, reducefunc, trainset, valset, dumpmodelname, terminatetraining=10, n_epochs=1 ):
    consec_badresult = 0
    bestloss = np.Inf
     
    model.epochs = 0  
    siz = 100
     
    for epoch in range(0,n_epochs):        
        tloss, vloss, tscore, vscore = 0 , 0, 0, 0 
        model.train()
        for ii, (useless, labels) in enumerate(trainset):
            print('ii'+str(ii))
            data2 = pickle.load(open("drive/MyDrive/sem6/mlass2/"+str(ii)+"batch_train.sav", "rb") )
            
            # print('target')
            # print(target)
            
            
            reducefunc.zero_grad()
             
            data2 = data2.view(data2.size()[0], data2.size()[1]*data2.size()[2]*data2.size()[3])
            generated = model.classifier(data2)
            del data2
            gc.collect()
            loss = lossfunc(generated, labels)
            # print(type(loss))
            loss.backward()
            reducefunc.step()
            tloss = tloss + (loss.item() * siz)  
            convert_prob_to_exactvalues(generated, labels)

             
            tscore += convert_prob_to_exactvalues(generated, labels) * siz 
        else:
            model.epochs += 1            
            with torch.no_grad():
                model.eval() 
                for jj, (useless, labels) in enumerate(abstrationval):
                    data2 = pickle.load(open("drive/MyDrive/sem6/mlass2/"+str(jj+432)+"batch_train.sav", "rb") ) 
                    data2 = data2.view(data2.size()[0], data2.size()[1]*data2.size()[2]*data2.size()[3])          
                    generated = model.classifier(data2)
                    del data2
                    gc.collect()
                    loss = lossfunc(generated, labels)
                    vloss = vloss + (loss.item() * siz)                    
                    vscore += convert_prob_to_exactvalues(generated, labels) * siz 
                     

                asize, bsize = len(trainset.dataset), len(valset.dataset)
                # extracting train and validation no. of samples
                tloss,  vloss, tscore, vscore  = tloss / asize,   vloss / bsize, tscore / asize,  vscore / bsize  
                # Dividing all 4 metrics to compute avg.

                 
               
                wandb.log({"Epoch": epoch,        
           "Train Loss": tloss,        
           "Train Acc": tscore,        
           "Valid Loss": vloss,        
           "Valid Acc": vscore})
                print("epoch"+str(epoch+1))
                print("training loss:"+str(tloss))
                print("Val loss:"+str(vloss))
                print("train accuracy:"+str(tscore))
                print("val accuracy:"+str(vscore))     
                if bestloss  > vloss:
                    
                    torch.save(model.state_dict(), dumpmodelname)
                    consec_badresult = 0
                    bestloss = vloss
                else:
                    consec_badresult += 1
                    if terminatetraining <= consec_badresult:
                        print("no more training needed")
                        model.load_state_dict(torch.load(dumpmodelname))
                        model.optimizer = reducefunc
                        return model
    model.optimizer = reducefunc
    return model
```

```python
# Whether to test on a gpu
from torch import cuda
trained_model = pickle.load(open("drive/MyDrive/sem6/mlass2/model.sav", "rb"))
 
gpuenabled =cuda.is_available()
print("find gpu"+str(cuda.is_available()))
if cuda.is_available():
    trained_model = trained_model.to('cuda')
```

```python
def yhat_acc(generated, label):
  useless3, yhat = torch.max(generated, dim=1) 
  acc = torch.mean(yhat.eq(label.data.view_as(yhat)).type(torch.FloatTensor))
  return acc.item(), yhat
```

```python
#  accuracy on test set
test_loader = abcd(testsamples, batch_size=collectiveset, shuffle=False, num_workers=2, pin_memory=True)
trained_model.eval()
test_acc = 0.0
pred_target = []
for jj, (data, label) in enumerate(test_loader):
  if gpuenabled:
    data, label = data.cuda(), label.cuda()
  generated = trained_model(data)
  

  accuracy, yhat = yhat_acc(generated, label) 
  # Multiply average accuracy times the number of examples
  pred_target.append([label,yhat])
  test_acc += accuracy * 100
test_acc = test_acc/len(test_loader.dataset)
print("test set accuracy:" + str(test_acc))
```

```python
#confusion matrix on test set
import numpy as np
confusionmatrix = np.zeros(shape=(10,10))
for i in pred_target:
  # print(i[0])
  for j in range(len(i[0])):
    actual = i[0][j].item()
    predicted = i[1][j].item()
    confusionmatrix[actual][predicted] += 1

print(confusionmatrix)
```

```python
#classwise confusion matrix and accuracy on test set
# print(np.sum(confusionmatrix))
for classwise in range(10):
  print("for class "+str(classwise))
  # print(confusionmatrix[classwise])
  acc = confusionmatrix[classwise][classwise] / np.sum(confusionmatrix[classwise])
  print("accuracy:"+str(acc) )
  mat = np.zeros(shape=(2,2))
  mat[0][0], mat[0][1] = confusionmatrix[classwise][classwise],  sum(confusionmatrix[classwise]) -  confusionmatrix[classwise][classwise]
  mat[1][0] = sum(confusionmatrix.T[classwise]) -  confusionmatrix[classwise][classwise]
  mat[1][1] = np.sum(confusionmatrix) - mat[0][0] - mat[0][1] - mat[1][0]
  print("confusion matrix")
  print(mat)
  
```
